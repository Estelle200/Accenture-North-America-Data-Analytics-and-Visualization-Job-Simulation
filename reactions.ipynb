### Cleaning the Reactions Dataset

```python
#Loading the csv for reactions and displaying the whole dataset.
import pandas as pd
reactions = pd.read_csv('C:/Users/Estar/Downloads/Reactions.csv')
display(reactions)
```

```python
#Removing irrelevant columns
reactions = reactions.drop(columns = ['Unnamed: 0', 'User ID'])
display(reactions)
```

```python
#removing null values 
reactions = reactions.dropna()
display(reactions)
```

```python
#removing duplicates
reactions = reactions.drop_duplicates()
display(reactions)
```

```python
#Renaming Columns
reactions.rename(columns = {'Type': 'Reaction_Type'}, inplace = True)
display(reactions)
```

```python
#This dataset is clean and ready to use.
reactions.to_csv('C:/Users/Estar/Downloads/cleaned_reactions.csv', index=False)
```
