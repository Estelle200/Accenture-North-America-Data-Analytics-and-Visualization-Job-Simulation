### Merging the Datasets for modelling

```python
import pandas as pd

#Load the cleaned DataFrames 
reactions = pd.read_csv('C:/Users/Estar/Downloads/cleaned_reactions.csv')         
content = pd.read_csv('C:/Users/Estar/Downloads/cleaned_content.csv')           
reaction_types = pd.read_csv('C:/Users/Estar/Downloads/cleaned_reaction_types.csv')
```

```python
#Merge reaction with content on 'Content ID'
df = pd.merge(reactions, content, on = 'Content ID', how = 'left')
display(df)
```

```python
#Merge the result with reaction_types on 'Reaction_Type'
reaction_summary = pd.merge(df, reaction_types, on =  'Reaction_Type', how = 'left')
display(reaction_summary)
```

```python
reaction_summary.columns = reaction_summary.columns.str.strip()
print(reaction_summary.dtypes)
```

```python
reaction_summary.rename(columns = {'Content ID': 'Content_ID'}, inplace = True)
display(reaction_summary)
```

```python
# Group by category and sum the scores

category_scores = reaction_summary.groupby('Category')['Score'].sum().reset_index()
category_scores = category_scores.sort_values(by='Score', ascending = False)
top_5_categories = category_scores.head(5)
display(category_scores)
display(top_5_categories)
```

```python
#This dataset is clean and ready to use.
reaction_summary.to_csv('C:/Users/Estar/Downloads/cleaned_reaction_summary.csv', index=False)
```

```python
category_scores.to_csv('C:/Users/Estar/Downloads/category_scores.csv', index=False)
```

```python
top_5_categories.to_csv('C:/Users/Estar/Downloads/top_5_categories.csv', index=False)
```
