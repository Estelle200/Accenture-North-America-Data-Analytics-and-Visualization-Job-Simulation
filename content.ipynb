### Cleaning the Content Dataset

```python
#Loading the csv for content and displaying the whole dataset.
import pandas as pd
content = pd.read_csv('C:/Users/Estar/Downloads/Content.csv')
display(content)
```

```python
#Removing irrelevant columns and renaming columns
#content = content.drop(columns = ['Unnamed: 0', 'User ID', 'URL'])
content.rename(columns = {'Type': 'Content_Type'}, inplace = True)
display(content)
```

```python
#Removing null values 
content = content.dropna()
display(content)
```

```python
#removing duplicates
content['Category'] = content['Category'].str.replace('"', '', regex=False)
content['Category'] = content['Category'].str.lower()
content = content.drop_duplicates()
display(content.head(20))
```

```python
#This dataset is clean and ready to use.
content.to_csv('C:/Users/Estar/Downloads/cleaned_content.csv', index=False)
```
